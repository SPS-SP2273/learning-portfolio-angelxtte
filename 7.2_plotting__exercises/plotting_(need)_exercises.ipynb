# Exercise 1

```python
from matplotlib import pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = 0.25 

plt.errorbar(x, y, xerr = x_err, yerr = y_err,
            color = "darkred", linestyle = "dashed",
            marker = "o", markerfacecolor = "yellow",
            ecolor = "black", capsize = 5,
            label = "Y values with errors")
# ecolor is error bar colour, marker is the type of marker for points
# capsize is the size of the caps at the end of the error bars
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("X vs Y")
plt.grid(alpha = 0.25)
plt.legend(loc = "lower right")
plt.show()
```

# Exercise 2

```python
import numpy as np
from matplotlib import pyplot as plt

data = np.linspace(0, 50, 1000)
y_value = (np.power(np.e, -data/10)) * np.sin(data)

plt.plot(data, y_value)
plt.xlabel("x", color = "darkorange", fontsize = 15)
plt.ylabel("y", color = "darkorange", fontsize = 15, rotation = 0)
plt.ylim([-1, 1])
plt.grid(alpha = 0.25)
plt.title(r"$y = e^{-x/10}\sin(x)$")
```
