# 1 A simple plot

## 1.1 Letâ€™s look at some code.

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
# Plotting the line itself
plt.xlabel('x-values') # X Label
plt.ylabel('y-values') # Y Label
plt.title('X vs Y') # Just the title
plt.grid(alpha=0.25) # Opacity of the Lines
plt.legend(loc='upper left') # Location of legend
plt.show() # Produces the graph as an output
````

### Things to note

There are some abbreviations that can be used for ```plt.plot()```:
|Long Form|Abbreviation|
|:---:|:---:|
|```colour```|```c```|
|```linestyle```|```ls```|
|```linewidth```|```lw```|

Hence, ```plt.plot(x, y, color = "red", linestyle = "dashed", linewidth = 2)``` and ```plt.plot(x, y, c = "red", ls = "dashed", lw = 2)``` produce the same results.

Jupyter notebook is an interactive environment so there will be an output even when ```plt.show()``` is not present, but it is good to include it so that it works in non-interactive environments too.

The plotting functions usually have default values for styling parameters so it can be kept simple too:


````python
plt.plot(x, y_1, y_2)
plt.show()
````

The order in which the keyword arguments are specified does not matter.

## 1.2 Adding another plot

````python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
# To add the second plot
plt.plot(x, y_2,
         label='Y2 values', color='blue')
plt.xlabel('x-values') # X Label
plt.ylabel('y-values') # Y Label
plt.title('X vs Y') # Just the title
plt.grid(alpha=0.25) # Opacity of the Lines
plt.legend(loc='upper left') # Location of legend
plt.show() # Produces the graph as an output
````

## 1.3 Yet another plot but with error bars

To add a plot with error bars, use ```errorbar()```:

````python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
# To add the second plot
plt.plot(x, y_2,
         label='Y2 values', color='blue')
# To add the third plot but with error bars
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values') # X Label
plt.ylabel('y-values') # Y Label
plt.title('X vs Y') # Just the title
plt.grid(alpha=0.25) # Opacity of the Lines
plt.legend(loc='upper left') # Location of legend
plt.show() # Produces the graph as an output
````

# 2 Better with NumPy

It is easier to use NumPy arrays instead of Python lists:

````python
import numpy as np

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)
np_y_3 = np.array(y_3)

plt.plot(np_x, np_y_1, label='Y values',
         color='red', linestyle='dashed')
# To add the second plot
plt.plot(np_x, np_y_2,
         label='Y2 values', color='blue')
# To add the third plot but with error bars
plt.errorbar(np_x, np_y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values') # X Label
plt.ylabel('y-values') # Y Label
plt.title('X vs Y') # Just the title
plt.grid(alpha=0.25) # Opacity of the Lines
plt.legend(loc='upper left') # Location of legend
plt.show() # Produces the graph as an output
````

## 2.1 Adding mathematical functions

One of the key advantages of NumPy arrays is that they allow for the generation of data-related mathematical functions easily.

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend(loc = "upper left")
plt.show()                                
````

Because $sin(x)$ lies between $\pm1$ but $x^2$ has no such bounds, the plot doesn't look very good. As such, another y-axis that shares the same x-axis must be added.

## We need another axis!

To plot multiple axes, ```twinx()``` and ```twiny()``` can be used. To have another y-axis that shares the same x-axis, ```twinx()``` is used:

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc = "lower left")

plt.twinx() # Share the same x-axis

plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend(loc = "lower right")
plt.show()    
````

To improve the plot due to the limited number of points present, ```np.linspace``` can be used:

````python
x = np.linspace(0, 6, 100)

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc = "lower left")

plt.twinx() # Share the same x-axis

plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend(loc = "lower right")
plt.show()    
````

```np.linspace(0, 6, 100)``` takes ```0``` as the start number, ```6``` as the end number, and ```100``` as the number of evenly spaced out numbers between ```0``` and ```6```.

### Things to note

There are two ```legend()``` calls, one for each axis.

When there are insufficient data points, the graph might look weird so ```np.linspace``` can be used to generate an array of data points that are evenly spaced out.

# 3 Saving to disc

To save the plot for use, use ```savefig()``` and specifiy the format with the extension of the file name (e.g. ```filename.pdf```) and the resolution using dots-per-inch (```dpi```).

````python
x = np.linspace(0, 6, 100)

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc = "lower left")

plt.twinx() # Share the same x-axis

plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend(loc = "lower right")
# plt.show()
# do not include plt.show() if saving the figure

plt.savefig('simple-plot.png', dpi=150)
````

# 4 A real example: Global Warming

## 4.1 Plotting data from files

````python
data = np.loadtxt("Complete_TAVG_daily.txt", skiprows = 24)
# np.loadtxt() is used to load data from a text file and return it as a NumPy array. delimiter is the string used to separate values in the file, which by default is a space.

date = data[:, 0] # All dates in an array
anomaly = data[:, -1] # All anomaly in an array

plt.plot(date, anomaly, alpha = 0.5)
plt.ylim([-8, 8]) # Limits the y-axis 
plt.show()
````

Adding a horizontal line at the zero value to highlight the trend shown by the data is done by using ```hlines()``` (or ```vlines()``` for vertical lines), which requires a $y$-value and starting and ending values for $x$.

````python
data = np.loadtxt("Complete_TAVG_daily.txt", skiprows = 24)
# np.loadtxt() is used to load data from a text file and return it as a NumPy array. delimiter is the string used to separate values in the file, which by default is a space.

date = data[:, 0] # All dates in an array
anomaly = data[:, -1] # All anomaly in an array

plt.plot(date, anomaly, alpha = 0.5)
plt.ylim([-8, 8]) # Limits the y-axis 
plt.hlines(0, date[0], date[-1], linestyle = "dashed", colors = "grey")
plt.show()
````

## 4.2 Styles

There are [standard style templates](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html), one of which is called **fivethirtyeight**:

````python
data = np.loadtxt("Complete_TAVG_daily.txt", skiprows = 24)
# np.loadtxt() is used to load data from a text file and return it as a NumPy array. delimiter is the string used to separate values in the file, which by default is a space.

date = data[:, 0] # All dates in an array
anomaly = data[:, -1] # All anomaly in an array

# Must be used at the top
plt.style.use("fivethirtyeight")

plt.plot(date, anomaly, alpha = 0.5)
plt.ylim([-8, 8]) # Limits the y-axis 
plt.hlines(0, date[0], date[-1], linestyle = "dashed", colors = "grey")
plt.xlabel("Date")
plt.ylabel("Temperature Anomaly")
plt.title("Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)")
plt.show()
````
