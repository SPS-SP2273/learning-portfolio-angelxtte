# Exercise 3

````python
squares = [i**2 for i in range(10)]
print(squares)
````

````python
%%timeit
squares = [i**2 for i in range(10)]
````

````python
squares = list()
for i in range(10):
    squares.append(i**2)
print(squares)
````

````python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
````

The first set of code's ```%%timeit``` output implies that from 7 runs of 1,000,000 loops each, it took a mean time of 868 ns, with a standard deviation of 26.1 ns, for the code to be executed. 

The second set of code's ```%%timeit``` output implies that from 7 runs of 1,000,000 loops each, it took a mean time of 1.1 s, with a standard deviation of 18.1 ns, for the code to be executed.

The results imply that the first set of code takes a shorter time to execute than the second set of code, which could be due to the fact that list comprehensions do not need to reference an already created list and append the integer to that list.
